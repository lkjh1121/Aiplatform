## 필요한 라이브러리 import

```python
import pandas as pd
import numpy as np
# 밀워키 팀의 경기 기록을 위한 라이브러리
from nba_api.stats.endpoints import leaguegamefinder
# 밀워키 팀과 상대 팀의 선수확인을 위한 라이브러리
from nba_api.stats.endpoints import BoxScoreUsageV2

# row 생략 없이 출력
pd.set_option('display.max_rows', 40)
# col 생략 없이 출력
pd.set_option('display.max_columns', None)
```

# 1. 밀워키 1610612749

```python
leaguegamefinder.LeagueGameFinder(team_id_nullable='1610612749')
```

```python
GSW_games = leaguegamefinder.LeagueGameFinder()
```

## 1.1. 18 19 시즌 경기

```python
df = GSW_games.get_data_frames()[0]
```

```python
df
```

## 1.2. 데이터에서 필요한 피처를 제외한 나머지 제거

```python
dsf = df[['GAME_ID','GAME_DATE','MATCHUP','WL']]
```

## 1.3. 데이터 저장에 필요한 리스트 

```python
gameid_list = dsf['GAME_ID'].to_list()
gamedate_list = dsf['GAME_DATE'].to_list()
gamematchup_list = dsf['MATCHUP'].to_list()
gamewl_list = dsf['WL'].to_list()
```

# 2. 매 경기의 선수들 확인

```python
for j,k,l in zip(gameid_list,gamedate_list,gamematchup_list):
    game = BoxScoreUsageV2(game_id=f'{j}')
    game = game.get_data_frames()[0][['GAME_ID','TEAM_CITY','PLAYER_NAME','MIN','START_POSITION']]
    game.to_csv(f'{k}{l}.csv')
```

```python
gameid_list = dsf['GAME_ID'].reset_index(drop=True)
gamedate_list = dsf['GAME_DATE'].reset_index(drop=True)
gamematchup_list = dsf['MATCHUP'].reset_index(drop=True)
gamewl_list = dsf['WL'].reset_index(drop=True)
```

```python
gameid_list.to_csv('gameid_list.csv',index=False)
gamedate_list.to_csv('gamedate_list.csv',index=False)
gamematchup_list.to_csv('gamematchup_list.csv',index=False)
gamewl_list.to_csv('gamewl_list.csv',index=False)
```
